#### 1) ¿Cómo medir el nivel de confianza de un modelo?
Las principales formas es usando las métricas de precission, recall y accuracy, además de poder usar otras métricas como el error cuadrático medio, o intervalos de confianza.

#### 2)¿Cuál es la utilidad de las técnicas: ¿Transfer learning y Fine Tuning?
El Transfer learning nos permite utilizar modelos previamente entrenados para poder facilitar el entrenamiento del nuevo modelo, así como optimizar su tiempo de ejecución.
Fine Tuning nos permirte hacer algo similar al transfer learning, reutilizando un modelo entrenado, pero esta vez cambiando el learning rate, y esto nos optimiza mucho tiempo de ejecución.

#### 3) ¿Para qué sirve la capa de MaxPooling?
Es una capa de una red convolucional que calcula el valor máximo del mapeo. Esto hace que resalte las características más revelantes dentro del ejemplo o imagen, para así optimizar tiempo de cómputo.

#### 4) ¿Qué parámetros requiere como mínimo la capa de Dropout en Keras?
El rate o porcentaje de perceptrones a los que se les aplica el drop, y la dimensión del vector.

#### 5) ¿Cuáles son los parámetros que aprender la red neuronal en las capas convolucionales?
Son los filtros que se usan en la capa, actualiza los pesos correspondientes dentro de cada filtro que se usa en esa capa.

#### 6) Regresión, Clasificación Binaria o Clasificación Multiclase ¿cuál es el problema más difícil de aprender para una red neuronal? Justifica tu respuesta

Es la Clasificación Multiclase, ya que ésta usa múltiples clasificaciones binarias incrustadas una dentro de otra, es decir se desenvuelve recursivamente, por lo que requiere mucho más trabajo de procesamiento.

#### 7) ¿Cuál de los siguientes métodos NO evita que un modelo se sobreajuste al conjunto de entrenamiento? Explica tu respuesta Early Stooping Dropout Data Augmentation Pooling
Pooling, pues éste método se utiliza para resaltar las careacterísticas principales de una muestra, después de haber aplicado una convolución, y al usarla en demasía, puede provocar que la red tienda a ciertas características sobre de otras.

Bagging, ya que ambas técnicas "quitan" una parte del conjunto que se está entrenando en la red.
